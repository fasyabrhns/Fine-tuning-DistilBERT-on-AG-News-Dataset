# Laporan Hasil Training: AG News

## Ringkasan Eksekutif
Model DistilBERT berhasil dilatih untuk mengklasifikasikan berita ke dalam 4 kategori (World, Sports, Business, Sci/Tech) dengan akurasi akhir mencapai **94.8%**.

## Metrik Evaluasi
Berdasarkan proses fine-tuning selama 3 epoch, berikut adalah detail performanya:

| Epoch | Training Loss | Validation Loss | Accuracy |
|-------|---------------|-----------------|----------|
| 1     | 0.1924        | 0.1782          | 93.96 %  |
| 2     | 0.1362        | 0.1661          | 94.53 %  |
| 3     | 0.1030        | 0.1683          | 94.80 %  |

## Analisis
- **Konvergensi:** Model belajar sangat cepat. Pada epoch pertama, akurasi sudah menembus 93%, menandakan bahwa *pretrained knowledge* dari DistilBERT sangat relevan dengan dataset berita ini.
- **Stabilitas:** Meskipun *Validation Loss* sedikit naik di epoch ke-3 (0.1683) dibanding epoch ke-2 (0.1661), akurasi tetap meningkat. Ini adalah tanda *early overfitting* yang sangat ringan, namun model masih berada di titik optimalnya.
- **Kesimpulan:** Dengan akurasi hampir 95%, model ini sudah sangat layak digunakan untuk sistem pengkategorian berita otomatis.
